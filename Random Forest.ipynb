```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sn
iris = load_iris()

df=pd.DataFrame(iris.data,columns=iris.feature_names)
df["target"] = iris.target
df["target_name"]=df.target.apply(lambda x:iris.target_names[x])
df
```

```python
iris.data
```

```python
sn.scatterplot(df["sepal length (cm)"],df["sepal width (cm)"],hue=df.target)
```

```python
sn.scatterplot(df["sepal length (cm)"],df["sepal width (cm)"],hue=df.target_name)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(["target_name","target"],axis=1),df.target,test_size=0.1)

from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10)
model.fit(X_train,y_train)
model.predict(X_test)
model.score(X_test,y_test)
```
